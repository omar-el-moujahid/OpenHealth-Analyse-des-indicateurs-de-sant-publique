```python
import pandas as pd 
```

```python
df_commun = pd.read_csv('Donnees/geographique/communes_france.csv', sep=';')
```

```python
df_commun.info()
```

```python
df_commun.head()
```

```python
df_commun[['code_commune','nom_comm']] = df_commun['code_commune,nom_comm'].str.split(',', n=1, expand=True)
```

```python
df_commun.head()
```

```python
df_commun.info()
```

```python
df_commun.drop(columns=['code_commune,nom_comm'], inplace=True)
```

```python
df_commun.head()
```

```python
##df_commun = df_commun.apply(lambda x: x.astype(str) if x.astype(str)==True )
df_commun['code_commune'] = pd.to_numeric(df_commun['code_commune'], errors='coerce')
```

```python
df_commun.dropna(inplace=True)
```

```python
df_commun['code_commune'] = df_commun['code_commune'].astype(int)
```

```python
df_commun.info()
```

```python
df_commun.head()
```

```python
df_commun.to_csv('Donnees/geographique/communes_france_clean.csv', sep=';', index=False)
```

```python
df_region = pd.read_csv('Donnees/geographique/region_france.csv', sep=';')
```

```python
df_region.info()
```

```python
df_region[['nom_region','code_region']] = df_region['nom,code'].str.split(',', expand=True)
```

```python
df_region.head()
```

```python
df_region.info()
```
